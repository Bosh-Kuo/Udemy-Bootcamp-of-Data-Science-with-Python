## Ex9_1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')
df.drop(columns = 'Id', inplace = True)
print(df)

sns.countplot(x = 'Species', data = df)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylabel('Frequency')
plt.ylim(0,80)
```

## Ex9_2

```python
sns.scatterplot(x = 'SepalLengthCm', 
                y = 'SepalWidthCm' , 
                data = df, 
                hue = 'Species', 
                alpha = 0.5)

plt.title('Relationship between Sepal Length and Sepal Width', fontsize = 12)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

sns.scatterplot(x = 'PetalLengthCm', 
                y = 'PetalWidthCm', 
                data = df, 
                hue = 'Species', 
                alpha = 0.5)

plt.title('Relationship between Petal Length and Petal Width', fontsize = 12)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

## Ex9_3

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, figsize = (10, 10))

sns.histplot(x = 'SepalLengthCm', data = df, ax = ax1)
ax1.set(title = 'Sepal Length histogram', ylabel = 'Frequency')

sns.histplot(x = 'SepalWidthCm', data = df, ax = ax2)
ax2.set(title = 'Sepal Width histogram', ylabel = 'Frequency')

sns.histplot(x = 'PetalLengthCm', data = df, ax = ax3)
ax3.set(title = 'Petal Length histogram', ylabel = 'Frequency')

sns.histplot(x = 'PetalWidthCm', data = df, ax = ax4)
ax4.set(title = 'Petal Width histogram', ylabel = 'Frequency')

plt.tight_layout(pad = 2)
```

## Ex9_4

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df)
plt.title('Box plot')
plt.show()
```

## Ex9_5

```python
corr = df.corr()  #Compute pairwise correlation of columns
sns.heatmap(corr, cmap = 'Blues', annot = True)

sns.pairplot(data = df)
plt.show()
```
