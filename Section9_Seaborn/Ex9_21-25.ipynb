```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
series = pd.read_csv( 'Airline_passengers.csv' , header=0, index_col=0, parse_dates=True,
squeeze=True)
```

## Ex9_21

```python
print(series.describe())

print('\nShape:', series.shape)
```

## Ex9_22

```python
series.plot(color = 'black')
plt.title('International airline passengers')
plt.grid(alpha = 0.5)
plt.xlabel('Time')
plt.ylabel('monthly totals in thousands')
plt.show()
```

## Ex9_23

```python
sns.boxplot(data = series)
plt.title('Boxplot')
plt.xlabel('Airline passengers')
plt.show()
```

## Ex9_24

```python
sns.histplot(series)
plt.title('Airline passengers - Histogram')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(series)
plt.title('Kernel density estimation');
```

## Ex9_25

```python
from pandas.plotting import lag_plot

lag_plot(series)
plt.grid(ls = '--', alpha = 0.5)
```
