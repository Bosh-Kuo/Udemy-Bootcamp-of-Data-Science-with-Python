## Ex9_6

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = sns.load_dataset(name = 'tips')
data.head(n = 5)
```

```python
print(data.describe())
print('\nColumns: ',data.columns)
print('\nShape: ',data.shape)
```

```python
# unique values
for column in data.columns[2:6]:
    print(column.title()+' Unique Values')
    print(data[column].unique())
    print()

#the proportion between men and women
print('\n', data['sex'].value_counts())

#the proportion between smokers and nonsmokers
print('\n', data['smoker'].value_counts())
```

## Ex9_7

```python
for variable in data.select_dtypes(include = 'category'):
    sns.countplot(x = variable, data = data)
    plt.title(variable.title()+' count')
    plt.show()
    
```

## Ex9_8

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = data)
plt.title('Relationship between total bill and tip')
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', data = data)
plt.title('Relationship between total bill and tip - by sex')
plt.show()
```

## Ex9_9

```python
sns.relplot(y = 'tip', x = 'size', hue = 'smoker', data = data)
plt.title('Relationship between tip and size - by smoker')
plt.show()
```

## Ex9_10

```python
sns.boxplot(data = data)
plt.title('Boxplot for quantitative variables')
plt.show()
```

```python
sns.histplot(data['tip'], kde = True)
plt.title('Distribution plot - tips')
plt.show()
```

```python
sns.histplot(data['total_bill'], kde = True)
plt.title('Distribution plot - Total bill')
plt.show()
```
