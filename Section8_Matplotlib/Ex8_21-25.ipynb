## EX8_21
- Present the descriptive statistics of the time series with the describe () function
- Show the shape of the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
series = pd.read_csv( 'Airline_passengers.csv' , header=0, index_col=0, parse_dates=True,
squeeze=True)
print(series)
print()
print(series.describe())
print("\n",series.shape)
```

## Ex8_22
- Make a line graph to demonstrate the evolution of the variable 'Airline-passengers' over time.

```python
series.plot(color = 'black')
plt.title('International airline passengers')
plt.grid(alpha = 0.5)
plt.xlabel('Time')
plt.ylabel('monthly totals in thousands')
plt.show()
```

## Ex8_23
- Make a boxplot characterize the variable 'Airline-passengers'.

```python
sns.boxplot(data = series)
plt.title('Boxplot')
plt.xlabel('Airline passengers')
plt.show()
```

## Ex8_24
- Make a histogram and a kde plot.

```python
sns.histplot(series)
plt.title('Airline passengers - Histogram')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(series)
plt.title('Kernel density estimation');
```

## Ex8_25
- Import the lag_plot function and apply the time series (this function generates a lag plot for time series.)

```python
from pandas.plotting import lag_plot
lag_plot(series)
plt.grid(ls = '--', alpha = 0.5)
```
