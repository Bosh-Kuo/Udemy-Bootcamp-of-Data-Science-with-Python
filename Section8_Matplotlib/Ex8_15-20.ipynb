## Ex8_15
- Load the attached Iris dataset in this lesson.

- Remove the Id column;

- Make a bar graph to show the frequency of the three Iris species in the dataset;

- Make a horizontal bar graph to obtain the previous result.



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')
print(df)
df.drop(columns = 'Id', inplace = True)
print(df)
frequency_iris = df['Species'].value_counts()

plt.bar(x = frequency_iris.index, height = frequency_iris)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylim(0,80)
plt.show()

plt.barh(y = frequency_iris.index,width = frequency_iris)
plt.title('Frequency of the three species in the Iris Dataset')
plt.xlim(0, 80)
plt.show()
```

## Ex8_16
- Make a pie chart to show the frequency of the three Iris species in the dataset;

```python
explode = (0, 0.1, 0.1)
labels = frequency_iris.index
plt.pie(frequency_iris, explode = explode, shadow = True, autopct = '%1.2f%%', labels = labels)
plt.title('Pie chart', fontsize = 15)
plt.legend(labels, loc = 'upper left')
plt.show()
```
