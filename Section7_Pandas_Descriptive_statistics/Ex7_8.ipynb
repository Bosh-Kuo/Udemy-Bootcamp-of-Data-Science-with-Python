```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('CardioGoodFitness.csv')
print(df)
```

- Generate a histogram for the age column;



```python
df['Age'].plot(kind = 'hist', bins = 30)
```

- Generate a histogram for all columns in the dataset;



```python
plt.figure()
df['Education'].plot(kind = 'hist', bins = 30)
plt.title("Education")

plt.figure()
df['Usage'].plot(kind = 'hist', bins = 30)
plt.title("Usage")

plt.figure()
df['Fitness'].plot(kind = 'hist', bins = 30)
plt.title("Fitness")

plt.figure()
df['Income'].plot(kind = 'hist', bins = 30)
plt.title("Income")

plt.figure()
df['Miles'].plot(kind = 'hist', bins = 30)
plt.title("Miles")
```

- Generate a pie chart for the gender column.



```python
fig, ax = plt.subplots() 
df.groupby(['Gender']).size().plot(kind='pie', title='Gender Ratio',textprops={'fontsize': 10},autopct='%.2f',  shadow=True)
ax.set_ylabel('Gender', size=22)
```
